# 4.**Subset Variables**   





* Select multipe columns with specific names

````python
df[['name','validate']]
````

* Select columns whose name matches regular expression regex.

````python
#select column where name end with k
df.filter(regex='k$')
````

* select the columns in positions (the first column is 0)

````python
# select the first and the third column in df
df.iloc[:,[0,3]]
````

* Select rows meeting logical condition, and only the specific columns.

````python
#select name and validate state where age is great than 20.
df.loc[df.age>20,['name','validate']]
````
